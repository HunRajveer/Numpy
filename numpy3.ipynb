##Numpy array operations

Lets first see how arrays behave like “vectors/matrices” when operations happen element-wise.

```python
# in numpy array is bit fast then python list and we can perform fast and quick operation , it is memory effecient
import numpy as np
arr = [1,2,3,4,5]
a = np.array(arr)
b = np.arange(5,10)
a
b
print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
print(b//a)
```

These things are already different from List and if we talk about one more special power it is **Broadcasting**

```python
# broadcasting is nothing but we will directly apply operation on our array or matrix
a +5
b-1
a**2
b//2
c = np.arange(1,13).reshape(3,4)
c
c*3
c//2
# so this is ow we can apply broadcasting in array
```

Even these things work with different shaped array the calculations will happen on every row. but dimensions must match

```python
# deep copy and shallow copy
#when we perform slicing in array and then broadcast or perfrom any other operation it will not affecgt the original array insted it will creat a copy inside the RAM
#eg
d = np.arange(1,15)
x = d[0:7]
x
```
